```python
import requests
import pandas as pd

# Define the API URL and API key
api_url = "https://browse.search.hereapi.com/v1/discover"
api_key = "iX8b5wL14DzJtbN5m5LlT9OzqHqN36arN5AtvEwdUd0"

# Initialize lists to store results
metro_results = []
resident_results = []

# Define the common parameters for the API requests
radius = 300  # You can adjust this radius as needed

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    # Extract the latitude and longitude from the DataFrame
    lat = row['Lat']
    lng = row['Lng']
    address = row['location']
    print(address)
    # Make API requests for "metrostation"
    params_metro = {
        "at": f"{lat},{lng}",
        "q": "metrostation",
        "apiKey": api_key,
        "r": radius
    }
    response_metro = requests.get(api_url, params=params_metro)

    # Make API requests for "store"
    params_resident = {
        "at": f"{lat},{lng}",
        "q": "apartment",
        "apiKey": api_key,
        "r": radius
    }
    response_resident = requests.get(api_url, params=params_resident)

    # Check if the requests were successful
    if response_resident.status_code == 200:
        resident_data = response_resident.json()
        resident_results.append(resident_data)

    if response_metro.status_code == 200:
        metro_data = response_metro.json()
        metro_results.append(metro_data)

# Now, metrostation_results and store_results contain the results for each row in the DataFrame
# You can process and analyze these results as needed
```

```python
# Initialize a list to store hospital counts
resident_count = []

# Iterate through the list of dictionaries
for i in range(len(resident_results)):
    if 'items' in resident_results[i] and isinstance(resident_results[i]['items'], list):
        # Initialize a count for hospitals within 300 meters
        count = 0
        for item in resident_results[i]['items']:
            if 'distance' in item and item['distance'] < 5000:
                count += 1
    else:
        count = 0

    resident_count.append(count)
```

```python
# Initialize a list to store metro station counts
metro_count = []

# Iterate through the list of dictionaries
for i in range(len(metro_results)):
    if 'items' in metro_results[i] and isinstance(metro_results[i]['items'], list):
        count = 0
        previous_station_distance = None  # To keep track of the previous station's distance

        for item in metro_results[i]['items']:
            if 'distance' in item and item['distance'] < 1500:
                # Check if the distance is greater than or equal to 100 meters compared to the previous station
                if previous_station_distance is None or item['distance'] - previous_station_distance >= 100:
                    count += 1
                    previous_station_distance = item['distance']
    else:
        count = 0

    metro_count.append(count)


# Now metro_count contains the counts of metro stations within 5000 meters, considering only one station if the distance between two stations is less than 100 meters.
```
